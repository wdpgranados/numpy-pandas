```python
import numpy as np

# 1. Matriz A: Precios de hoy (en soles)
# [Leche, Pan]
# [Arroz, Fideos]
precios_hoy = np.array([[4.50, 1.20],
                        [3.80, 2.50]])

# 2. Matriz B: Cuánto subieron hoy (inflación o ajuste)
incrementos = np.array([[0.50, 0.10],
                        [0.20, 0.30]])

# 3. La magia de NumPy: Sumamos todo de un solo golpe
precios_mañana = precios_hoy + incrementos

print("Precios actualizados para mañana:")
print(precios_mañana)
```

```python
# 1. Matriz A: Precios de hoy (en soles)
# [Leche, Pan]
# [Arroz, Fideos]
precios_hoy = np.array([[4.0,1.50],
                       [4.20,2.50]])
incrementos = np.array([[0.50,0.20],
                        [0.20,0.30]])
subida_precios=precios_hoy + incrementos

print("precio anterior:\n",precios_hoy)
print("precio actual: \n",subida_precios)
```

```python
import pandas as pd

# 1. Creamos nuestra "hoja de cálculo"
# Aquí ponemos los nombres de los productos y sus precios actuales
datos = {
    'Producto': ['Leche', 'Pan', 'Arroz', 'Fideos'],
    'Precio_Hoy': [4.50, 1.20, 3.80, 2.50],
    'Incremento': [0.50, 0.10, 0.20, 0.30]
}

# Convertimos los datos en un DataFrame de Pandas
inventario = pd.DataFrame(datos)

# 2. Creamos una nueva columna calculando el precio de mañana
# Pandas nos permite usar los nombres de las columnas directamente
inventario['Precio_Mañana'] = inventario['Precio_Hoy'] + inventario['Incremento']

# 3. Mostramos el resultado
print("--- Inventario Actualizado ---")
print(inventario)
```
