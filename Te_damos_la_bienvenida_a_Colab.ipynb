```
import numpy as np

# Creamos dos listas de números (vectores) [cite: 61, 62]
pesas_izquierdas = np.array([5, 10, 15])
pesas_derechas = np.array([2, 2, 2])

# Sumamos ambas listas directamente [cite: 64]
total_por_brazo = pesas_izquierdas + pesas_derechas

print(total_por_brazo)
# Resultado: [ 7, 12, 17]
```

```
import pandas as pd

# Creamos una tabla (DataFrame) [cite: 116, 198]
datos = {
    'Producto': ['Pan', 'Leche', 'Huevos', 'Manzana'],
    'Precio': [1.50, 4.20, 0.50, 2.10]
}

df = pd.DataFrame(datos)

# Filtramos los que cuestan más de 2 soles [cite: 122, 200]
caros = df[df['Precio'] > 2]

print(caros)
```
